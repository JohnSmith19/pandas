### 시간 범위기 주기 설정하기
시간 범위를 인덱스로 지정하면 DatetimeIndex 자료형이 만들어 집니다. 그리고 freq 속성값을 지정하면 시간 간격을 조절하여 DatetimeIndex를 만들 수 있다.

- B: 평일만 포함
- C: 사용자가 정의한 평일만 포함
- D: 달력 일자 단위
- W: 주간 단위
- M: 월 마지막 날만 포함
- SM: 15일과 월 마지막 날만 포함
- BM: M 주기의 값이 휴일이면 제외하고 평일만 포함
- CBM: BM 에 사용자 정의 평일을 적용
- MS: 월 시작일만 포함
- SMS: 월 시작일과 15일만 포함
- BMS: MS 주기의 값이 휴일이면 제외하고 평일만 포함
- CBMS: BMS에 사용자 정의 평일을 적용
- Q: 3, 6, 9, 12월 분기 마지막 날만 포함
- BQ: 3, 6, 9, 12월 분기 마지막 날이 휴일이면 제외하고 평일만 포함
- QS: 3, 6, 9, 12월 분기 시작일만 포함
- BQS: 3, 6, 9, 12월 분기 시작일이 휴일이면 제외하고 평일만 포함
- A: 년의 마지막 날만 포함
- BA: 년의 마지막 날이 휴일이면 제외하고 평일만 포함
- AS: 년의 시작일만 포함
- BAS: 년의 시작일이 휴일이면 제외하고 평일만 포함
- BH: 평일을 시간 단위로 포함(09:00 ~ 16:00)
- H: 시간단위로 포함(00:00 ~ 00:00)
- T: 분 단위 포함
- S: 초 단위 포함
- L: 밀리초 단위 포함
- U: 마이크로초 단위 포함
- N: 나노초 단위 포함

```python
# Date 열의 자료형을 datetime Oject로 변환하여 ebola DataFrame을 다시 생성한다.
ebola = pd.read_csv('../data/country_timeseries.csv', parse_dates=['Date'])

ebola.head().iloc[:, :5]
```

```python
# Date 열을 인덱스로 지정한후 Date 열의 최대 최소값으로 시간 범위를 생성, new_idx 에 저장한다.
ebola.index = ebola['Date']
new_idx = pd.date_range(ebola.index.min(), ebola.index.max())
new_idx
```

```python
# 시간 순서를 맞추기 위해 reversed 메서드를 사용해 인덱스를 뒤집는다.
new_idx = reversed(new_idx)
```

```python
# reindex를 사용하여 새로 생성한 인덱스를 새로운 인덱스로 지정한다.
# 2015년 01월 01일 테이터와 같은 아예 없던 날짜가 추가된다.
ebola = ebola.reindex(new_idx)
ebola.head().iloc[:, :5]
```

```python
ebola.tail().iloc[:, :5]
```

```python
# 각 나라의 에볼라 발병일 옮기기
last_valid = ebola.apply(pd.Series.last_valid_index)
last_valid
```

```python
first_valid = ebola.apply(pd.Series.first_valid_index)
first_valid
```

```python
# 각 나라의 에볼라 발병일을 동일한 출발선으로 옮기려면 에볼라가 가장 처음 발병한 날에서 각 나라의 에볼라 
# 발병일을 뺀 만큼 옮긴다.

earliest_date = ebola.index.min()
earliest_date
```

```python
shift_values = last_valid - earliest_date
shift_values
```

```python
# 에볼라 최초 발병일을 기준으로 각 나라의 에볼라 발병일을 옮긴다.
ebola_dict = {}
for idx, col in enumerate(ebola):
    d = shift_values[idx].days
    shifted = ebola[col].shift(d)
    ebola_dict[col] = shifted
    
ebola_shfit = pd.DataFrame(ebola_dict)
ebola_shfit.tail()
```

```python
# 인덱스를 Day 열로 지정하고 그래프에 필요없는 Date, Day 열을 삭제한다.
ebola_shfit.index = ebola_shfit['Day']
ebola_shfit = ebola_shfit.drop(['Date', 'Day'], axis=1)

ebola_shfit.tail()
```

```python
fig, ax = plt.subplots()
ax = ebola_shfit.iloc[:,:].plot(ax=ax)
ax.legend(fontsize=7, loc=2, borderaxespad=0.)
plt.show()
```
